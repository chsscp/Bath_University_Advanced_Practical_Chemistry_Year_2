```python
import numpy as np
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
Temperature = np.array([100, 200, 300, 400, 500, 600, 700, 800, 900, 1000, 1100, 1200, 1300, 1400, 1500])
Diffusion = np.array([3.47578797e-04, 3.47578797e-04, 1.71670606e-03, 4.13489708e-03, 2.18762447e-02, 1.36015747e-01, 
                      2.90576568e-01, 6.93472254e-02, 1.08966402e-01, 2.51636797e-01, 5.03405716e-01, 7.22862928e-01,
                      9.18047160e-01, 1.02884720e+00, 9.54932975e-01])
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress((1 / Temperature)[2:], (np.log(Diffusion))[2:])
```

```python
Ea = ((-slope * 8.314) / 1000) / 96.485
```

```python
print(Ea)
```
